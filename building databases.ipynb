```python
import pandas as pd
import numpy as np
```

```python
not_hot_data = pd.read_csv('spotify_millsongdata.csv') 
#reading the file, provide your own path, depending on where you saved the file.
```

```python
not_hot_data
```

```python
#dropping the columns we don`t need and saving our new DF as 'not_hot_list'
not_hot_list = not_hot_data.drop(columns=['link', 'text'])
not_hot_list
```

```python
#switching the order of the columns, so that it matches the order of our hot songs DF

not_hot_list = not_hot_list[['song', 'artist']]
not_hot_list
```

```python
from random import sample
```

```python
list_of_numbers = np.arange(0, 57650, 1).tolist()
```

```python
random_sample=sample(list_of_numbers,2500)
```

```python
reduced=not_hot_list.iloc[random_sample].reset_index(drop=True)
```

```python
reduced
```

```python
hot_songs = pd.read_csv('../Morning/hot_songs.csv') 
hot_songs
```

```python
#=()[]{}
merged = pd.concat([reduced, hot_songs], axis=0).reset_index(drop=True)
```
