```python
#=()[]{}
def search_song(title, artist, limit):
    list_of_ids = []
    try:
        results = sp.search(q="tracks: "+title+" artist: "+artist, limit=limit)
        for i in range (0, limit):
            id = results['tracks']['items'][i]['id'] 
            list_of_ids.append(id)
    except:
        print("Song not found!")
        list_of_ids.append("")
    return list_of_ids
```

```python
id = []
X=0
import time
for i in range (0, hot_songs.shape[0]):
    list_of_ids = search_song(hot_songs.iloc[i, 0], hot_songs.iloc[i, 1], 1)
    id.append(list_of_ids[0])
    X=X+1
    print(X)
    time.sleep(5)
    
```

```python
def get_audio_features(list_of_song_ids):
    
    n=50
    final = [list_of_song_ids[i * n:(i + 1) * n] for i in range((len(list_of_song_ids) + n - 1) // n )]

    audio_features = []
    for index, subset in enumerate(final):
        print("Getting the uadio features of all the songs in the subset,", index+1)
        list_of_dict = sp.audio_features(subset) # [{},{},{}] # {"keys":[]}
        audio_features += list_fo_dict  
        time.sleep(30)
    
    return pd.DataFrame(audio_features)
```

```python
try_1 = get_audio_features(id)
```

```python
#=()[]{}
def add_audio_features(df, audio_features_df):
    table = df.merge(audio_features_df, how="left", left_on = "id", right_on = "id").drop_duplicates().reset_index()
    return table
```

```python
try_2 = add_audio_features(hot_songs, try_1)
```
